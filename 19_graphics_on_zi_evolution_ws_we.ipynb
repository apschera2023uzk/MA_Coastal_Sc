```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', size=12) 
plt.style.use('seaborn-poster')

outputdir = "/home/qwertz/Dokumente/N_eigene_Dokumente/Geo_Met_MSc/Semester_4_5/Thesis_prep_Juni_23/MA/master_plots/19_graphics/"

steps = 6
D = 1.00000000025
xs = np.linspace(0, 2000, steps)
xs2 = np.linspace(0+(2000/(steps+1)), 2000-(2000/(steps+1)), steps-1)
zs = np.linspace(0, 2000, steps)
xss, zss = np.meshgrid(xs, zs)
us = np.tile(np.zeros([len(xs)]), len(xs))
vs = np.tile(- zs*5, (steps,1)).T
vss = vs
print(vs)

plt.figure(figsize=(15,8))
plt.title("Temporal development of boundary layer height")
plt.quiver(xss, zss, us, vss, color="red", label="subsidence $w_s$")#, headlength=3)# , scale=10)
plt.quiver(xs2, [1000]*len(xs2), [0]*len(xs2), [1000]*len(xs2), color="green", label="entrainment $w_e$")# , scale=10)
plt.axhline(1000, color="black", label="$z_i$")
plt.ylabel("height [m]")
plt.legend(loc="lower left")
plt.xticks([])
plt.savefig(outputdir+"bl_development.png")

#################################
# Fge eine Wolke hinzu
# Fge Ozeanoberflche unten hinzu
# Fge entrainment Wirbel hinzu

# Plane auch animierte Version fr MA defense
```
